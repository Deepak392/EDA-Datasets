```python
# Fill missing numeric columns with median
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
df.isnull().sum()
```

```python
df = df.rename({'Overall ratting':'Overall rating'}, axis = 1)
```

```python
df['Culture & Values'] = df['Culture & Values'].fillna('NaN')
df['Diversity & Inclusion'] = df['Diversity & Inclusion'].fillna('NaN')
df['Career Opportunities'] = df['Career Opportunities'].fillna(0)
df['Compensation and Benefits'] = df['Compensation and Benefits'].fillna(0)
df['Senior Management'] = df['Senior Management'].fillna(0)
df['CEO Approval'] = df['CEO Approval'].fillna('NaN')
df['Recommended'] = df['Recommended'].fillna('NaN')


df.isnull().sum()
```
